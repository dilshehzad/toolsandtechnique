```python
#load csv file
dengue_train_data = pd.read_csv(r'E:\MS Data Science\Semester 1\Tools and Techniques for Data Science\Project\imd\finalprojecttt\origianal data\dengue_features_train.csv')
```

```python
dengue_train_data['city'] = dengue_train_data['city'].astype('category') 
dengue_train_data['city'] = dengue_train_data['city'].cat.codes

dengue_train_data['week_start_date'] = dengue_train_data['week_start_date'].astype('category') 
dengue_train_data['week_start_date'] = dengue_train_data['week_start_date'].cat.codes
```

```python
dengue_train_data_correct = dengue_train_data.fillna(dengue_train_data.mean()) 
dengue_train_data_correct.head()
```

```python
dengue_train_data_correct.isnull().sum()
```

```python
y = dengue_train_data_correct['total_cases']
dengue_train_data_correct_features = dengue_train_data_correct.drop(['total_cases'], 1)
X = dengue_train_data_correct_features
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error, r2_score

#test_train_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2852, random_state=101)
#logreg
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)

y_pred = regr.predict(X_test)


# The coefficients
print('Coefficients: \n', regr.coef_)
# The mean squared error
print("Mean absoulute error: %.2f"
      % mean_absolute_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))
```
